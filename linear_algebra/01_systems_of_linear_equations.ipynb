### System of Equations as Lines

#### Material

**Notes**

- $m = \frac{\Delta_{y}}{\Delta_{x}} $ (slope)
- $b = y - mx$ (y-intercept)

**Linear Equation → Line**

- $x + y = 10$

````python
x_values = [12,10,8,6,4,2,0,-2]
y_values = [-2,0,2,4,6,8,10,12]

m = (y_values[-1] - y_values[0]) / (x_values[-1] - x_values[0])
b = y_values[0] - m * x_values[0]

plt.plot(x_values, y_values, label='Line')

for i, (x, y) in enumerate(zip(x_values, y_values)):
    plt.text(x, y, f'({x}, {y})', ha='left', va='bottom')

legend = f'(Slope={m}, Y-intercept={b})'
plt.legend([legend])

plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting a Line in Cartesian Coordinates')
plt.show()
````

- $x + 2y = 12$

````python
x_values = [12,8,4,0,-12]
y_values = [0,2,4,6,12]

m = (y_values[-1] - y_values[0]) / (x_values[-1] - x_values[0])
b = y_values[0] - m * x_values[0]

plt.plot(x_values, y_values, label='Line')

for i, (x, y) in enumerate(zip(x_values, y_values)):
    plt.text(x, y, f'({x}, {y})', ha='left', va='bottom')

legend = f'(Slope={m}, Y-intercept={b})'
plt.legend([legend])

plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting a Line in Cartesian Coordinates')
plt.show()
````

**Linear dependence between rows**

Linearly independent

```bash
[[1, 1],
 [1, 2]] # No row is a multiple of the other one
```

Linearly dependent

```bash
[[1, 1],
 [2, 2]] # Second row is a multiple of the first row
```

**The Determinant**

Matrix is **Singular** if

\begin{align}
[a][b] * k &= [c][d] \\
ak &= c \\
bk &= d \\
\frac{c}{a} &= \frac{d}{b} = k \\
ad &= cb \\
ad - cb &= 0 \: (Determinant)
\end{align}

````python
def determinant(M):
    r, c = M.shape
    
    if r == 2 and c == 2:
        return M[0][0] * M[1][1] - M[1][0] * M[0][1]
    else:
        raise ValueError
````

````python
linearly_independent_matrix2D = np.array([[1,1],[1,2]])
````

````python
determinant(linearly_independent_matrix2D)
````

````python
linearly_dependent_matrix2D = np.array([[1,1],[2,2]])
````

````python
determinant(linearly_dependent_matrix2D)
````

Matrix is singular ⟺ Determinant is zero

### System of Equations as Planes (3x3)

#### Material

**Linear equation in 3 variables -> Plane**

- $x + y + z = 0$
- $x + 2y + z = 0$
- $x + y + 2z = 0$

````python
# Define a range for a and b
x_range = np.linspace(-10, 10, 100)
y_range = np.linspace(-10, 10, 100)

# Create a meshgrid for a and b
x, y = np.meshgrid(x_range, y_range)

# Calculate c values based on the given equations
z1 = -(x + y)
z2 = -(x + 2 * y)
z3 = -(x + y) / 2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surfaces defined by the three equations
ax.plot_surface(x, y, z1, alpha=0.5, facecolors='r', rstride=100, cstride=100, label='x + y + z = 0')
ax.plot_surface(x, y, z2, alpha=0.5, facecolors='g', rstride=100, cstride=100, label='x + 2y + z = 0')
ax.plot_surface(x, y, z3, alpha=0.5, facecolors='b', rstride=100, cstride=100, label='x + y + 2z = 0')

# Find the point of intersection
intersection_point = np.linalg.solve([[1, 1, 1], [1, 2, 1], [1, 1, 2]], [0, 0, 0])
ax.scatter(*intersection_point, color='black', s=5)

# Set labels
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')

# Show the plot
plt.legend()
plt.show()
````

### System of Equations as Matrices (3x3)

#### Material

**Linear dependence and independence**

- $x + y + z = 0$
- $2x + 2y + 2z = 0$
- $3x + 3y + 3z = 0$

Row 1 + Row 2 = Row 3 (Linearly Dependent)

- $x + y + z = 0$
- $x + y + 2z = 0$
- $x + y + 3z = 0$

(Row 1 + Row 3) / 2 = Row 2 (Linearly Dependent)

- $x + y + z = 0$
- $x + 2y + z = 0$
- $x + y + 2z = 0$

No relations between rows (Linearly Independent)

**The determinant (3x3)**

````python
def determinant(M):
    r, c = M.shape
    
    if r == 2 and c == 2:
        return M[0][0] * M[1][1] - M[1][0] * M[0][1]
    elif r == 3 and c == 3:
        return ( 
            ((M[0][0] * M[1][1] * M[2][2]) + (M[0][1] * M[1][2] * M[2][0]) + (M[0][2] * M[1][0] * M[2][1])) -
            ((M[0][2] * M[1][1] * M[2][0]) + (M[0][0] * M[1][2] * M[2][1]) + (M[0][1] * M[1][0] * M[2][2]))
        )
    else:
        raise ValueError
````

````python
linearly_independent_matrix3D = np.array([[1,1,1], [1,2,1], [1,1,2]])
````

````python
determinant(linearly_independent_matrix3D)
````

### Keypoints

- Linearly Independent Vectors (Full Rank Matrix):
    - The system of linear equations has a unique solution.
    - The determinant of the matrix is non-zero.
    - The matrix is invertible.

- Linearly Dependent Vectors (Rank-Deficient Matrix):
    - The system of linear equations may have either no solution or infinite solutions.
    - The determinant of the matrix is zero.
    - The matrix is not invertible.

- Useful Library
    - np.linalg.solve
